```python
print("\nMissing values before cleaning:\n", df.isnull().sum())

# Example: Fill numeric columns with median, categorical with mode
for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values after cleaning:\n", df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print("\nDuplicate rows found:", duplicates)

df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
# Try converting any column with "date" in its name to datetime
for col in df.columns:
    if "date" in col.lower():df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

```
